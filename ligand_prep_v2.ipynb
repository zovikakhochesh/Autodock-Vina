```python
from rdkit import Chem
from rdkit.Chem import AllChem
from meeko import MoleculePreparation
from meeko import PDBQTWriterLegacy
import os

# для запуска укажите относительный путь файла в input(), например - data/example.sdf или data/example.smi

input_file = input()
if input_file.endswith('.sdf'):
    os.makedirs("output_2/output_sdf", exist_ok=True)
    i = 0
    for mol in Chem.SDMolSupplier(input_file):
        if mol.HasProp("_Name"): 
            name = mol.GetProp("_Name") 
        else:
            i += 1
            name = f"unknown_{i}"
        mol = Chem.AddHs(mol) 
        AllChem.EmbedMolecule(mol) 
        mk_prep = MoleculePreparation(min_ring_size=6) 
        molsetup_list = mk_prep(mol) 
        molsetup = molsetup_list[0] 
        pdbqt_string = PDBQTWriterLegacy.write_string(molsetup)[0]
        with open(f"output_2/output_sdf/{name}.pdbqt", 'w') as f:
                f.write(pdbqt_string)
        
elif input_file.endswith('.smi'):
    os.makedirs("output_2/output_smi", exist_ok=True)
    i = 0
    for mol in Chem.SmilesMolSupplier(input_file, delimiter=' ', titleLine=True):
        if mol.HasProp("_Name"): 
            name = mol.GetProp("_Name") 
        else:
            i += 1
            name = f"unknown_{i}" 
        mol = Chem.AddHs(mol) 
        AllChem.EmbedMolecule(mol) 
        mk_prep = MoleculePreparation(min_ring_size=6) 
        molsetup_list = mk_prep(mol) 
        molsetup = molsetup_list[0] 
        pdbqt_string = PDBQTWriterLegacy.write_string(molsetup)[0]
        with open(f"output_2/output_smi/{name}.pdbqt", 'w') as f:
                f.write(pdbqt_string)
else:
    print('unknown file format')
```
