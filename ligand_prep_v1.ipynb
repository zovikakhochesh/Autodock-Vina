```python
# подготовка sdf файла
from rdkit import Chem
from rdkit.Chem import AllChem
from meeko import MoleculePreparation
from meeko import PDBQTWriterLegacy
import os

mols_sdf_list = []
name_sdf_list = []
for mol in Chem.SDMolSupplier('data/example.sdf'): # считываем молекулы sdf
    mols_sdf_list.append(mol) # добавляем молекулу в список
    if mol.HasProp("_Name"): # проверяем есть ли имя у молекулы
        name = mol.GetProp("_Name") # извлекаем свойство имени при прохождении проверки
    else:
        name = "unknown"
    name_sdf_list.append(name) # добавляем имя в список

for i in range(len(mols_sdf_list)): # для каждой молекулы в списке: 
    mol = mols_sdf_list[i] # создаём локальную переменную 
    mol = Chem.AddHs(mol) # добавляем атомы водорода
    AllChem.EmbedMolecule(mol) # создаём 3d конформер
    mols_sdf_list[i] = mol # изменяем молекулу в списке
```

```python
# подготовка smi файла
mols_smi_list = []
name_smi_list = []
for mol in Chem.SmilesMolSupplier('data/example.smi', delimiter=' ', titleLine=True):  # считываем молекулы smi
    mols_smi_list.append(mol) # добавляем молекулу в список
    if mol.HasProp("_Name"): # проверяем есть ли имя у молекулы
        name = mol.GetProp("_Name") # извлекаем свойство имени при прохождении проверки
    else:
        name = "unknown"
    name_smi_list.append(name) # добавляем имя в список
        
for i in range(len(mols_smi_list)): # для каждой молекулы в списке: 
    mol = mols_smi_list[i] # создаём локальную переменную 
    mol = Chem.AddHs(mol) # добавляем атомы водорода
    AllChem.EmbedMolecule(mol) # создаём 3d конформер
    mols_smi_list[i] = mol # изменяем молекулу в списке
```

```python
# генерация PDBQT из sdf
os.makedirs("output/output_sdf", exist_ok=True) # создамим папку для выходных данных

for i in range(len(mols_sdf_list)): # для каждого элемента из двух списков (длина списка молекул и списка имён равны):
    name = name_sdf_list[i] # создаём локальную переменную для имени
    mol = mols_sdf_list[i] # создаём локальную переменную для молекулы RDKit
    mk_prep = MoleculePreparation() # создаём функцию парамеризации
    molsetup_list = mk_prep(mol) # применяем параметризацию к молекуле, создаём список объектов MoleculeSetup (1 объект в списке)
    molsetup = molsetup_list[0] # извлекаем из списка единственный объект
    pdbqt_string = PDBQTWriterLegacy.write_string(molsetup)[0] # создаем из объекта MoleculeSetup строку PDBQT
    with open(f"output/output_sdf/{name}.pdbqt", 'w') as f: # записываем полученную строку в файл в созданной папке
            f.write(pdbqt_string)  
```

```python
# генерация PDBQT из smi
os.makedirs("output/output_smi", exist_ok=True) # создамим папку для выходных данных

for i in range(len(mols_smi_list)): # для каждого элемента из двух списков (длина списка молекул и списка имён равны):
    name = name_smi_list[i] # создаём локальную переменную для имени
    mol = mols_smi_list[i] # создаём локальную переменную для молекулы RDKit
    mk_prep = MoleculePreparation() # создаём функцию парамеризации
    molsetup_list = mk_prep(mol) # применяем параметризацию к молекуле, создаём список объектов MoleculeSetup (1 объект в списке)
    molsetup = molsetup_list[0] # извлекаем из списка единственный объект
    pdbqt_string = PDBQTWriterLegacy.write_string(molsetup)[0] # создаем из объекта MoleculeSetup строку PDBQT
    with open(f"output/output_smi/{name}.pdbqt", 'w') as f: # записываем полученную строку в файл в созданной папке
            f.write(pdbqt_string)  
```
